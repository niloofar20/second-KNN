```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
from sklearn import preprocessing
```

```python
df = pd.read_csv('/content/drug200.csv')
df.head()
```

```python
X = df[['Age', 'Sex', 'BP', 'Cholesterol', 'Na_to_K']].values
X[0:5]
Y = df['Drug']
Y[0:5]
```

```python
from sklearn.preprocessing import LabelEncoder
le_sex = LabelEncoder()
le_BP = LabelEncoder()
le_Chol = LabelEncoder()
le_sex.fit(['F','M'])
X[:,1] = le_sex.transform(X[:,1])
le_BP.fit([ 'LOW', 'NORMAL', 'HIGH'])
X[:,2] = le_BP.transform(X[:,2])
le_Chol.fit([ 'NORMAL', 'HIGH'])
X[:,3] = le_Chol.transform(X[:,3])
X[0:5]
scalar = preprocessing.StandardScaler()
scalar.fit(X)
X_std = scalar.transform(X)
X_std[0:5]
```

```python
le_Y = LabelEncoder()
le_Y.fit(['drugA', 'drugB', 'drugC', 'drugX', 'drugY'])
Y = le_Y.transform(Y)
Y[0:5]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X_std, Y, test_size=0.2, random_state=3)
```

```python
x_train_scale = scalar.fit_transform(X_train)
x_test_scale = scalar.transform(X_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
my_modle = DecisionTreeClassifier(random_state=0)
my_modle.fit(x_train_scale, Y_train)
y_predict = my_modle.predict(x_test_scale)

```

```python
from sklearn.metrics import accuracy_score
accuracy_score(Y_test, y_predict)
```

```python
from sklearn.neighbors import KNeighborsClassifier
my_modle = KNeighborsClassifier(n_neighbors=5)
my_modle.fit(x_train_scale, Y_train)
y_predict = my_modle.predict(x_test_scale)
accuracy_score(Y_test, y_predict)

```
